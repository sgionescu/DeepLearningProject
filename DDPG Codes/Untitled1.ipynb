```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
q_max_lr0_01 = pickle.load(open("q_max_mountain_car_100episodes_critic_lr0_01.p", "rb"))
q_max_lr0_001 = pickle.load(open("q_max_mountain_car_100episodes_critic_lr0_001.p", "rb"))
q_max_lr0_0005 = pickle.load(open("q_max_mountain_car_100episodes_critic_lr0_0005.p", "rb"))
#print(q_max)

# actor

q_max_actor_lr0_001 = pickle.load(open("q_max_mountain_car_100episodes_critic_lr0_0005_actor_lr0_001.p", "rb"))
q_max_actor_lr0_0001 = pickle.load(open("q_max_mountain_car_100episodes_critic_lr0_0005.p", "rb"))
q_max_actor_lr0_01 = pickle.load(open("q_max_mountain_car_100episodes_critic_lr0_0005_actor_lr0_01.p", "rb"))

def get_mean_var(reward_hist):
    ts = pd.Series(reward_hist)
    
    return pd.rolling_mean(ts, 5)

qmax1 = get_mean_var(q_max_lr0_01)
qmax2 = get_mean_var(q_max_lr0_001)
qmax3 = get_mean_var(q_max_lr0_0005)

qmax1_actor = get_mean_var(q_max_actor_lr0_01)
qmax2_actor = get_mean_var(q_max_actor_lr0_001)
qmax3_actor = get_mean_var(q_max_actor_lr0_0001)



#ts3 = pd.Series(q_max_lr0_0005)

#fig = plt.figure()
#plt.plot(qmax1, label='Critic lr =0.01, Actor lr = 0.0001')
#plt.plot(qmax2, label='Critic lr = 0.001, Actor lr = 0.0001')
#plt.plot(qmax3, label='Critic lr = 0.0005, Actor lr = 0.0001')
#plt.plot(pd.rolling_mean(ts3, 20))
#plt.xlabel('Episode Number')
#plt.ylabel('Max Q-value')

#plt.legend(loc='best', bbox_to_anchor=(1, 0.35), fontsize=12) 
#plt.title('Mountain Car Critic Results',fontsize=16)

#plt.show()
#fig.savefig('Rewards_bipedal_critic_lr0_01_1000episodes.jpg')



fig,axe = plt.subplots(2,1,figsize=(12,12))


axe[0].plot(qmax1, label='Critic lr =0.01, Actor lr = 0.0001')
axe[0].plot(qmax2, label='Critic lr = 0.001, Actor lr = 0.0001')
axe[0].plot(qmax3, label='Critic lr = 0.0005, Actor lr = 0.0001')
axe[0].plot([103]*len(qmax1),'k--')
axe[0].set_ylabel('Max Q-value', fontsize=20)
axe[0].set_title('Mountain Car Critic Results',fontsize=20)
axe[0].legend(loc='best', bbox_to_anchor=(1, 1),fontsize=11) 
axe[0].legend(loc='best', bbox_to_anchor=(1, 0.28),fontsize=11) 
axe[0].set_xlabel("Episodes", fontsize=16)

#axe[1].set_ylim(0,40)
#axe[1].set_xlim(0,59000)
axe[1].plot([103]*len(qmax1_actor),'k--')
axe[1].plot(qmax1_actor, label='Critic lr =0.0005, Actor lr = 0.01')
axe[1].plot(qmax2_actor, label='Critic lr =0.0005, Actor lr = 0.001')
axe[1].plot(qmax3_actor, label='Critic lr =0.0005, Actor lr = 0.0001')

axe[1].legend(loc='best', bbox_to_anchor=(1, 0.28),fontsize=11) 
axe[1].set_title('Mountain Car Actor Results',fontsize=20)
axe[1].set_ylabel("Max Q-value", fontsize=20)


plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.xlabel("Episodes", fontsize=16)
plt.show()
fig.savefig('Mountain Car Results.jpg')
```
