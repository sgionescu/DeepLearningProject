```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rewards = pickle.load(open("rewards_mountain_car_100episodes_critic_lr0_01.p", "rb"))
ts = pd.Series(rewards)

fig = plt.figure()
plt.plot(pd.rolling_mean(ts, 10))
plt.xlabel('Episode Number')
plt.ylabel('Rewards')
plt.show()
fig.savefig('Rewards_mountain car_100episodes_critic_lr0_01.jpg')
```
